```python
import requests
import lxml
from bs4 import BeautifulSoup
```

# Crawling Experiments

```python
# access url
def getPage(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) '
                      'Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE '
    }
    f = requests.get(url, headers=headers)
    return f
```

```python
# parse page
def parse(page):
    soup = BeautifulSoup(page.content, 'lxml')
    return soup
```

```python
 # create url to crawl
url = "https://www.rottentomatoes.com/top/bestofrt/"
page = getPage(url)
page_content = parse(page)
```

```python
page_content
```

```python
# getting anchor tags in a table and parsing them out
movies = page_content.find('div',{'class':'discovery-grids-container'}).find_all('a')
print(movies)
```

# Crawl top movies Rotten Tomatos

```python
# #create list of urls
movies_lst = []

# fill list of urls
num = 0
for anchor in movies:
    urls = 'https://www.rottentomatoes.com' + anchor['href']
    movies_lst.append(urls)
    num +=1
    movie_url = urls
    movie_f = getPage(movie_url)
    movie_soup = BeautifulSoup(movie_f.content, 'lxml')
    movie_content = movie_soup.find('div', {
      'class': 'movie_synopsis clamp clamp-6 js-clamp'
    })
    print(num, urls, '\n', 'Movie:' + anchor.find('span',{'class':'p--small'}).string.strip())
    print('Movie info:' + movie_content.string.strip())
```
